```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import PIL
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers,models
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
mobile = tf.keras.applications.mobilenet.MobileNet()
```

```python
IMAGE_SHAPE =  (256, 256)

training_path = "C:/Users/Unmeet Patel/Documents/Non Academic/Lung cancer detection project/dataset/training"
testing_path = "C:/Users/Unmeet Patel/Documents/Non Academic/Lung cancer detection project/dataset/test"
validation_path = "C:/Users/Unmeet Patel/Documents/Non Academic/Lung cancer detection project/dataset/validation"

image_generator =  tf.keras.preprocessing.image.ImageDataGenerator(rescale=1/255, rotation_range=20, width_shift_range=0.2, height_shift_range=0.2, shear_range=0.2, zoom_range=0.2, horizontal_flip=True, vertical_flip=True, fill_mode='nearest')

training_data = image_generator.flow_from_directory(training_path, target_size = IMAGE_SHAPE, batch_size = 32)
testing_data = image_generator.flow_from_directory(testing_path, target_size = IMAGE_SHAPE, batch_size = 32)
validation_data = image_generator.flow_from_directory(validation_path, target_size = IMAGE_SHAPE, batch_size = 32, shuffle=False)
```

```python
training_data[0]
```

```python
testing_data[0]
```

```python
for layers in mobile.layers:
  layers.trainable = False
```

```python
num_of_disease = 3

model = Sequential()

for layer in mobile.layers[:-1]:
    model.add(layer)

model.add(Dense(num_of_disease, activation='softmax'))
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001),
                     loss='categorical_crossentropy',
                     metrics=['accuracy'])
```

```python
model.summary()
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping()

history = model.fit(training_data,
                    epochs=30,
                    validation_data=validation_data,
                    batch_size=32,
                    callbacks=[early_stopping],
                    verbose=1)
```

```python
model.save('model1.h5')
```

```python
model.save('model1.keras')
```

```python
scores = model.evaluate(testing_data)
scores
```
